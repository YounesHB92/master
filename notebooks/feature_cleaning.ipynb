# loading modules

```python
from src.utils import load_env_variables
working_dir, env = load_env_variables()

import os
os.chdir(working_dir)

import pandas as pd

```

# loading feature files

```python
features_path = os.getenv("FEATURES_DIR")
feature_file = "features.csv"
data = pd.read_csv(os.path.join(features_path, feature_file))
data.head(10)
```

# feature cleaning
## drop unnecessary columns

```python
to_delete = [
    "coords",
    "image.1",
    "image_convex",
    "image_filled",
    "label",
    "moments_normalized-0-0",
    "moments_normalized-0-1",
    "moments_normalized-1-0",
    "slice"
]

data_dropped = data.drop(columns=to_delete, axis=1)
```

## getting first member and turning into floats

```python
for column in data_dropped.columns:
    data_dropped[column] = data_dropped[column].apply(lambda x: x.replace("[", "").replace("]", ""))
    if column == "image":
        pass
    else:
        data_dropped[column] = data_dropped[column].apply(lambda x: float(x))

data_clean = data_dropped.copy()
```

## check if we have any NaN values

```python
nan_info = data_clean.isnull().sum()
nan_features = nan_info[nan_info > 0].index
if nan_features.empty:
    print("No NaN values found in the features.")
else:
    print(f"NaN values found in the following features: {nan_features}")
```

## saving clean features

```python
name = "features_clean.csv"
data_clean.to_csv(os.path.join(features_path, name), index=False)
```
