# Loading configs

## loading all the config files we  have

```python
checkpoints_dir = os.getenv('CHECKPOINTS_DIR')
config_files = [file for file in os.listdir(checkpoints_dir) if file.endswith('.yaml')]
print("Total config files found: ", len(config_files))
```

## getting information from the config files

```python
info = {}
for config_file in config_files:
    config_path = os.path.join(checkpoints_dir, config_file)
    config_info = utils.get_config_info(config_path)
    info[config_file] = config_info
```
